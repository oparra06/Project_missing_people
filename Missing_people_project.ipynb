```python
from plotnine import *
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Desaparecidos.csv")
df.head(10)
```

```python
df = df.query("Municipio == 'VISTAHERMOSA' or Municipio == 'URIBE' or Municipio == 'LA MACARENA' or Municipio == 'MESETAS' or Municipio == 'SAN VICENTE DEL CAGUÁN'" )
df.Municipio.unique()
```

```python
df = df[df.Anho != 0]
```

```python
df.Anho.value_counts().sort_index(ascending=False)
```

```python
pd.pivot_table(data=df, index['Anho'])
```

```python
missing_people = df.groupby(['Anho','Municipio']).size().reset_index(name='count')
```

```python
missing_people.head(80)
```

```python
chart = alt.Chart(missing_people).mark_line().encode(
    x='Anho:O',
    y='count',
    color='Municipio'

)
chart.save('municipios_desaparecidos.html')
```

```python
df_missing = pd.read_csv("Desaparecidos_1000.csv")
df_missing['Desaparecidos_1000'] = df_missing['Desaparecidos_1000'].str.replace(",", ".").astype(float)
df_missing
```

```python
chart_des_1000= alt.Chart(df_missing).mark_circle().encode(
    x='Desaparecidos:Q',
    y='Desaparecidos_1000',
    color='Mun_Caguan',
    tooltip=['Municipio', 'Desaparecidos_1000']
)
chart_des_1000.save('desaparecidos1000.html')
```
